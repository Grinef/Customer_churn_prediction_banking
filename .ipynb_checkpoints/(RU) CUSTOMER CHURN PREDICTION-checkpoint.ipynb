## Построим BASELINE-Модель

```python
def gini_coefficient(y_true: pd.Series,
                     y_scores: pd.Series) -> float:
    """
    Вычисляет коэффициент Джини на основе ROC AUC.
    
    :param y_true: Истинные метки классов
    :param y_scores: Предсказанные вероятности
    :return: Коэффициент Джини
    """
    auc_score = roc_auc_score(y_true, y_scores)
    return 2 * auc_score - 1

def plot_roc_curve(y_true: pd.Series,
                   y_scores: pd.Series, ax) -> None:
    """
    Строит ROC-кривую на заданной оси.
    
    :param y_true: Истинные метки классов
    :param y_scores: Предсказанные вероятности
    :param ax: Ось для построения графика
    """
    fpr, tpr, _ = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)
    ax.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})')
    ax.plot([0, 1], [0, 1], 'k--', label='No Skill')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('ROC-кривая')
    ax.legend()

def plot_confusion_matrix(y_true: pd.Series,
                          y_pred: pd.Series, ax) -> None:
    """
    Строит матрицу путаницы на заданной оси.
    
    :param y_true: Истинные метки классов
    :param y_pred: Предсказанные метки классов
    :param ax: Ось для построения графика
    """
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(ax=ax, cmap=plt.cm.Blues)
    ax.set_title('Матрица ошибок')

def evaluate_model(
    model,
    X_train: pd.DataFrame,
    y_train: pd.Series,
    X_val: pd.DataFrame,
    y_val: pd.Series,
    save_name: str
) -> None:
    """
    Оценивает производительность модели на тренировочных и валидационных данных.
    
    :param model: Обученная модель
    :param X_train: Тренировочные данные
    :param y_train: Истинные метки тренировочных данных
    :param X_val: Валидационные данные
    :param y_val: Истинные метки валидационных данных
    """
    # Предсказания
    pred_train = model.predict(X_train)
    pred_val = model.predict(X_val)
    pred_proba_train = model.predict_proba(X_train)[:, 1]
    pred_proba_val = model.predict_proba(X_val)[:, 1]

    # Метрики F1
    f1_train = f1_score(y_train, pred_train)
    f1_val = f1_score(y_val, pred_val)
    relative_change_f1 = (f1_val - f1_train) / f1_train
    
    # ROC-AUC и Gini
    auc_train = roc_auc_score(y_train, pred_proba_train)
    auc_val = roc_auc_score(y_val, pred_proba_val)
    gini_train = gini_coefficient(y_train, pred_proba_train)
    gini_val = gini_coefficient(y_val, pred_proba_val)
    relative_change_gini = (gini_val - gini_train) / gini_train
    
    
    # Таблица результатов
    results = pd.DataFrame({
        'Sample': ['Train', 'Val', 'Относительное отколнение'],
        'GINI': [gini_train, gini_val, relative_change_gini],
        'F1': [f1_train, f1_val, relative_change_f1]
    })
    display(results.round(2))

    # Построение графиков
    fig, ax = plt.subplots(1, 2, figsize=(14, 6))
    
    plot_roc_curve(y_val, pred_proba_val, ax[0])
    plot_confusion_matrix(y_val, pred_val, ax[1])

    plt.tight_layout()
    plt.savefig(save_name)
    plt.show()

# Пример использования функции
baseline_model = LogisticRegression(random_state=RANDOM_STATE, class_weight='balanced')
baseline_model.fit(X_train, y_train)

evaluate_model(baseline_model, X_train, y_train, X_val, y_val, save_name='ROC_AUC_base_Lasso')
```

**Мы виидим, что на валидационных данных качество выросло, это может говорить о недообучение Модели. Возможно, мы построили слишком лёгкую Модель для данной задачи. Попробуем альтернативу в виде случайного леса.**

## Случайный лес

```python
model_rf = RandomForestClassifier(random_state=RANDOM_STATE,
                                  class_weight='balanced')

model_rf.fit(X_train, y_train)

evaluate_model(model_rf, X_train, y_train, X_val, y_val, save_name='ROC_AUC_RandomForest')
```

*Теперь мы однозначно можем сказать, что случайный лес решает нашу задачу, однако, мы сталкиваемся с проблемой переобучения. Чтобы избавиться от этого, попробуем Подобрать гиперпараметры.*


## Случайный лес подбор гипер-параметров

```python
# Определение гиперпараметров для поиска
params = {
    'n_estimators': [20, 50, 70, 100, 150, 175, 200, 225],
    'max_depth': [2, 4, 6],
    'bootstrap': [True, False],
    'min_samples_split': [4, 6, 8, 10, 12, 14]
}

# Создание модели и настройка GridSearchCV
clf_rf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=RANDOM_STATE, class_weight='balanced'),
    param_grid=params,
    cv=4,
    n_jobs=-1,
    scoring=make_scorer(f1_score)
)

# Обучение модели с подбором гиперпараметров
clf_rf.fit(X_train, y_train)
```

```python
best_model = clf_rf.best_estimator_
print(best_model)
```

```python
best_model.fit(X_train, y_train)

evaluate_model(best_model, X_train, y_train, X_val, y_val, save_name='ROC_AUC_RF_best')
```

**Теперь относительное отколнение приемлемо и Модель эффективно справляется с ранжированием классов. Можем дообучить Модель на всей выборке для разработки и проверить её качество на тестовых данных**

```python
best_model_dev = clf_rf.best_estimator_

X_dev = pd.concat([X_train, X_val])
y_dev = pd.concat([y_train, y_val]) 

best_model_dev.fit(X_dev, y_dev)
```

## Тестирование модели

**На завершающем этапе необходимо протестировать отобранную модель.**

```python
evaluate_model(best_model_dev, X_dev, y_dev, X_test, y_test, save_name='ROC_AUC_FINALLY_Model')
```
