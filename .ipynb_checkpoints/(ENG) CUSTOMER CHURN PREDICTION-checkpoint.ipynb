# Customer Churn

Clients have started leaving "Beta-Bank." Every month. Not many, but noticeably. The bank's marketing team calculated that retaining current customers is cheaper than acquiring new ones.

The task is to predict whether a customer will leave the bank in the near future. You have been provided with historical data on customer behavior and account terminations.

Build a model with the highest possible F1 score. To successfully complete the project, the F1 score needs to reach at least 0.59. Check the F1 score on the test set independently.

Additionally, measure the AUC-ROC and compare its value with the F1 score.

## Data Splitting

```python
target = df['exited'] 
features = df.drop(columns=['exited'], axis=1) 
```

```python
X_train, X_tech, y_train, y_tech = train_test_split(
    features, target, test_size=0.4, random_state=RANDOM_STATE, stratify=data['exited'])

X_val, X_test, y_val, y_test = train_test_split(
    X_tech, y_tech, test_size=0.5, random_state=RANDOM_STATE, stratify=y_tech) 

print('Train sample size:', "{0:.0%}".format(X_train.shape[0]/features.shape[0]))
print('Out-of-sample size:', "{0:.0%}".format(X_val.shape[0]/features.shape[0]))
print('Test smaple size:', "{0:.0%}".format(X_test.shape[0]/features.shape[0]))
```

```python
numeric = ['credit_score', 'age', 'tenure', 'balance', 'num_of_products', 'estimated_salary']
```

```python
scaler = StandardScaler()
scaler.fit(X_train[numeric])

X_train[numeric] = scaler.transform(X_train[numeric])
X_val[numeric] = scaler.transform(X_val[numeric])
X_test[numeric] = scaler.transform(X_test[numeric])
```

## Create BASELINE-MODEL

```python
def gini_coefficient(y_true: pd.Series,
                     y_scores: pd.Series) -> float:
    """
    Calculates the Gini coefficient based on ROC AUC.
    
    :param y_true: True class labels
    :param y_scores: Predicted probabilities
    :return: Gini coefficient
    """
    auc_score = roc_auc_score(y_true, y_scores)
    return 2 * auc_score - 1

def plot_roc_curve(y_true: pd.Series,
                   y_scores: pd.Series, ax) -> None:
    """
    Plots the ROC curve on the given axis.
    
    :param y_true: True class labels
    :param y_scores: Predicted probabilities
    :param ax: Axis for plotting the graph
    """
    fpr, tpr, _ = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)
    ax.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})')
    ax.plot([0, 1], [0, 1], 'k--', label='No Skill')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('ROC Curve')
    ax.legend()

def plot_confusion_matrix(y_true: pd.Series,
                          y_pred: pd.Series, ax) -> None:
    """
    Plots the confusion matrix on the given axis.
    
    :param y_true: True class labels
    :param y_pred: Predicted class labels
    :param ax: Axis for plotting the graph
    """
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(ax=ax, cmap=plt.cm.Blues)
    ax.set_title('Confusion Matrix')

def evaluate_model(
    model,
    X_train: pd.DataFrame,
    y_train: pd.Series,
    X_val: pd.DataFrame,
    y_val: pd.Series,
    save_name: str
) -> None:
    """
    Evaluates the model's performance on training and validation data.
    
    :param model: Trained model
    :param X_train: Training data
    :param y_train: True labels for training data
    :param X_val: Validation data
    :param y_val: True labels for validation data
    """
    # Predictions
    pred_train = model.predict(X_train)
    pred_val = model.predict(X_val)
    pred_proba_train = model.predict_proba(X_train)[:, 1]
    pred_proba_val = model.predict_proba(X_val)[:, 1]

    # F1 Metrics
    f1_train = f1_score(y_train, pred_train)
    f1_val = f1_score(y_val, pred_val)
    relative_change_f1 = (f1_val - f1_train) / f1_train
    
    # ROC-AUC and Gini
    auc_train = roc_auc_score(y_train, pred_proba_train)
    auc_val = roc_auc_score(y_val, pred_proba_val)
    gini_train = gini_coefficient(y_train, pred_proba_train)
    gini_val = gini_coefficient(y_val, pred_proba_val)
    relative_change_gini = (gini_val - gini_train) / gini_train
    
    # Results Table
    results = pd.DataFrame({
        'Sample': ['Train', 'Val', 'Relative Change'],
        'GINI': [gini_train, gini_val, relative_change_gini],
        'F1': [f1_train, f1_val, relative_change_f1]
    })
    display(results.round(2))

    # Plotting
    fig, ax = plt.subplots(1, 2, figsize=(14, 6))
    
    plot_roc_curve(y_val, pred_proba_val, ax[0])
    plot_confusion_matrix(y_val, pred_val, ax[1])

    plt.tight_layout()
    plt.savefig(save_name)
    plt.show()

# Example usage of the function
baseline_model = LogisticRegression(random_state=RANDOM_STATE, class_weight='balanced')
baseline_model.fit(X_train, y_train)

evaluate_model(baseline_model, X_train, y_train, X_val, y_val, save_name='ROC_AUC_base_Lasso')
```

**We see that the performance on the validation data has improved, which may indicate that the model is underfitting. It is possible that we have built a model that is too simple for this task. Let's try an alternative in the form of a Random Forest.**

## Random Forest

```python
model_rf = RandomForestClassifier(random_state=RANDOM_STATE,
                                  class_weight='balanced')

model_rf.fit(X_train, y_train)

evaluate_model(model_rf, X_train, y_train, X_val, y_val, save_name='ROC_AUC_RandomForest')
```

*Now we can definitively say that the Random Forest is addressing our problem, but we are facing overfitting issues. To resolve this, we will try tuning the hyperparameters.*


## Random Forest Hyperparameter Tuning

```python
# Define hyperparameters for search
params = {
    'n_estimators': [20, 50, 70, 100, 150, 175, 200, 225],
    'max_depth': [2, 4, 6],
    'bootstrap': [True, False],
    'min_samples_split': [4, 6, 8, 10, 12, 14]
}

# Create model and configure GridSearchCV
clf_rf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=RANDOM_STATE, class_weight='balanced'),
    param_grid=params,
    cv=4,
    n_jobs=-1,
    scoring=make_scorer(f1_score)
)

clf_rf.fit(X_train, y_train)
```

```python
best_model = clf_rf.best_estimator_
print(best_model)
```

```python
best_model.fit(X_train, y_train)

evaluate_model(best_model, X_train, y_train, X_val, y_val, save_name='ROC_AUC_RF_best')
```

**Now the relative change is acceptable, and the model is effectively handling class ranking. We can retrain the model on the entire development dataset and evaluate its performance on the test data.**

```python
best_model_dev = clf_rf.best_estimator_

X_dev = pd.concat([X_train, X_val])
y_dev = pd.concat([y_train, y_val]) 

best_model_dev.fit(X_dev, y_dev)
```

## Testing Model

```python
evaluate_model(best_model_dev, X_dev, y_dev, X_test, y_test, save_name='ROC_AUC_FINALLY_Model')
```

- Missing values were present in the "tenure" field (number of years the customer has been with the bank). After analysis, these missing values were replaced with "-1".
